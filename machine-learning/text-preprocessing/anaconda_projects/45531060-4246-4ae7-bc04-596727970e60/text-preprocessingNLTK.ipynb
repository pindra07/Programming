```python
words = ["eating", "eats","eaten", "writing", "writes", "programming", "programs", "running", "runner", "history", "finally", "finalized", "writing", "Writes"]
```

# Porter Stemming

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"----" + stemming.stem(word))
```
